2. Before we start getting information from the API, let's take a look at the different types of air quality measures that are available to us and their definition. Query the API to get all the different possible **parameters** and their definitions.

```python

l_res=[]
retour=200
indic=1
while retour==200  :
    url = f"https://api.openaq.org/v3/parameters/{indic}"
    headers = { 'X-API-Key': 'a680fc227081ae5f19a820a25e32cf59f975baffb047ae65c7cf985e405a4d10' }
    response = requests.get(url=url,headers=headers)
    if ('results' in response.json()) :
        l_res.append(response.json()['results'][0])
    retour=response.status_code
    indic = indic+1

```
