# House Market

Let's go a little further into python.

1. Download the four files “superficie.csv” / “nombre_de_sdb.csv” / “nombre_de_chambres.csv” / "house_price.csv"

2. Read the files and create the 4 corresponding Dataframes


```python
import pandas as pd
```

```python
df_surface=pd.read_csv('house_market/superficie.csv', names=['id','surface'])
df_bathroom=pd.read_csv('house_market/nombre_de_sdb.csv', names=['id','bathrooms'])
df_rooms=pd.read_csv('house_market/nombre_de_chambres.csv', names=['id','rooms'])
df_price=pd.read_csv('house_market/house_price.csv',index_col=0 )
```

```python
display(df_surface.head(4))
display(df_bathroom.head(4))
display(df_rooms.head(4))
display(df_price.head(4))
```

3. Give a column name to the three datasets that don't have a header in the source file

4. Make an inner join of these four datasets into a single dataframe.

```python
df_1=pd.merge(left=df_rooms, right=df_surface, how="inner", on ='id')
df_2=pd.merge(left=df_bathroom, right=df_price, how="inner", on ='id')
df_data=pd.merge(left=df_1, right=df_2, how="inner", on ='id')
```

```python
display(df_data)
```

5. What is the average surface area of the houses in our dataset?

```python
df_data['surface'].mean()
#df_data.groupby('id')['surface'].sum().mean()
```

6. What is the median and average number of rooms?

```python
display(df_data['rooms'].mean())
display(df_data['rooms'].median())
#display(df_data.groupby('id')['rooms'].sum().agg(['mean','median']))
```

7. What is the average cost of a house?

```python
df_data['house_price'].mean()
```

8. What is the average cost of a house, depending on the number of rooms it has?

```python
df_data.groupby('rooms')['house_price'].mean()
```

10. The average cost per room doesn't tell us much. Let's try to categorize by size.

  - Create a new column in your dataset that we'll call home_size
Create five categories that respectively correspond to:
      - "very large" $=$ "a house larger than 25,000 sqrt_feet"
      - "large" $=$ "a house between 20,000 and 25,000 sqrt_feet"
      - "medium" $=$ "a house between 15,000 and 20,000 sqrt_feet"
      - "small" $=$ "a house between 10,000 and 15,000 sqrt_feet"
      - "very small" $=$ a house less than 10,000 sqrt_feet"

Fill in the corresponding rows in the house_size column.

```python
def calc_size(x) :
    
    if x < 10000 :
        l_size="very small"
    elif x < 15000 :
        l_size="small"
    elif x < 20000 : 
        l_size="medium"
    elif x < 25000 : 
        l_size="large"
    else : 
        l_size="very lage"
    return l_size
df_data["home_size"]=df_data['surface'].apply(calc_size)

display(df_data)
    
```

11. What is the average cost of a house depending on its size category?


```python
df_data.groupby('home_size')['house_price'].mean()
```
